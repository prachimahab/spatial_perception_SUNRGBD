# Remove identifying information from raw data files

```python
import os
import pandas as pd
import csv
```

```python
src = '/Users/prachimahableshwarkar/Documents/GW/FacialAge/FacialAge_MTurk/BNav_EC2/DepthDuration/v2_depth_duration_MTurk/data'
dest = '/Users/prachimahableshwarkar/Documents/GW/Depth_MTurk/cleaned_spatial_perception_SUNRGBD/numerical_estimate/data/data-files/'
```

```python
for file in os.listdir(src):
    if '.csv' in file:
        dpath = src + '/' + file
        df = pd.read_csv(dpath, index_col=None, header=0)
        df.drop('workerId', inplace=True, axis=1)
        # random subject id
        subjID = str(df.subjID.unique()[0])
        df.to_csv(dest + subjID + '.csv')
        
        
        
```
