```python
import numpy as np
import pandas as pd
from scipy.stats import stats
from sklearn.linear_model import LinearRegression
import random
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
from time import process_time
import csv
import matplotlib.colors
```

```python
matrix_5000 = pd.read_csv('/Users/pmahableshwarkar/Documents/Depth_Project/spatial_perception_SUNRGBD/joint_analyses/avg_randSplitMatrix_n5000.csv')
# matrix_5000 = pd.read_csv('/Users/prachimahableshwarkar/Documents/GW/Depth_MTurk/spatial_perception_SUNRGBD/joint_analyses/avg_randSplitMatrix_n5000.csv')

mean_df_corr = matrix_5000.set_index('Unnamed: 0')
```

```python
bottom_mean_df_corr = mean_df_corr.tail(12)
bottom_mean_df_corr = bottom_mean_df_corr.drop(columns = ['r1_Discrim_PC_125', 'r1_Discrim_PC_250','r1_Discrim_PC_1000','r1_VE_PC_125', 'r1_VE_PC_250', 'r1_VE_PC_1000',
                                    'r1_Discrim 125 RT', 'r1_Discrim 250 RT','r1_Discrim 1000 RT', 'r1_VE 125 RT', 'r1_VE 250 RT', 'r1_VE 1000 RT'])
```

```python
bottom_mean_df_corr
```

```python
import matplotlib.colors

norm = matplotlib.colors.Normalize(-1,1)
colors = [[norm(-1.0), "aqua"],
          [norm(-0.2), "blue"],
          [norm(0), "black"],
          [norm( 0.2), "red"],
          [norm( 1.0), "yellow"]]


cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", colors)


plt.figure(figsize=[17,17])

ax = sns.heatmap(bottom_mean_df_corr, annot=True, fmt=".2f", cmap= cmap, square=True, center=0, vmin=-1, vmax=1, edgecolors='w', linewidths=3, cbar_kws={"shrink": .82}, annot_kws={'fontsize': 12, 'color': 'white'}) #notation: "annot" not "annote"
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)


plt.savefig('rand_5000.png', dpi=300)

plt.show();
```

```python
# Modify the normalization range from -0.7 to 0.7
norm = matplotlib.colors.Normalize(-0.7, 0.7)

colors = [[norm(-0.7), "aqua"],
          [norm(-0.2), "blue"],
          [norm(0), "black"],
          [norm(0.2), "red"],
          [norm(0.7), "yellow"]]

cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", colors)

plt.figure(figsize=[17, 17])

ax = sns.heatmap(bottom_mean_df_corr, annot=True, fmt=".2f", cmap=cmap, square=True, center=0, vmin=-0.7, vmax=0.7, edgecolors='w', linewidths=3, cbar_kws={"shrink": .82}, annot_kws={'fontsize': 18})
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.savefig('ODC_rand_5000.png', dpi=300)
plt.show();
```
