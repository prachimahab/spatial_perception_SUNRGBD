```python
from numpy import random
import matplotlib.pyplot as plt
import matplotlib
import json

#generates the colored mask grid given the path of the output file
def plotMask(path):
    data = random.random((50,68)) #(row, col)
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']

    random.shuffle(colors) #randomly shuffles the order of the colors 
    #print(colors, "Shuffled")
    my_cmap = matplotlib.colors.ListedColormap(colors) #creates custom color map for the plot
    
    plt.figure(figsize =(4.5,7))
    
    img = plt.imshow(data ,interpolation='nearest', cmap=my_cmap)
    plt.axis('off')
    #dpi sets the resolution
    plt.savefig(path, bbox_inches='tight', dpi = 200)

#generates all the versions of the mask where total is the number of masks 
#generate 256 (total number of trials per experiment)
#generate json file for the paths of all masks (for js folder in experiment)
def generateMasks_Json(total):
    json_data = []
    for i in range(total):
        img_dict = dict()
        path =  "/Users/prachi/Documents/depth_duration/mturk_code/depth_duration_MTurk/masks/" +"mask_" + str(i) + ".jpg"
        img_dict["mask_path"] = "masks/" +  "mask_" + str(i) + ".jpg"
        img_dict["id"] = "mask_" + str(i)
        json_data.append(img_dict)
        plotMask(path)
    print(json_data)
    with open("masks.json", "w") as outfile:
        json.dump(json_data, outfile, indent = 4)



generateMasks_Json(256)

        



#Need to check that this mask works in task.js
#Then need to generate 256
#need to figure out how to randomly assign mask image in the task.js code (json file?)
```
